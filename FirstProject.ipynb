# Data Analysis and Visualizations

```python
cancelled_perc = df['is_canceled'].value_counts(normalize = True)
print(cancelled_perc)

plt.figure(figsize = (5,4))
plt.title ('Rerservation status count')
plt.bar(['Not canceled', 'Canceled'], df['is_canceled'].value_counts(), edgecolor = 'k', width = 0.7)
plt.show
```

```python
plt.figure(figsize = (8,4))
ax1 = sns.countplot(x= 'hotel', hue = 'is_canceled', data = df, palette = 'Blues')
legend_labels,_ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation status in different hotels', size = 20)
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.legend(['not_canceled', 'canceled'])
plt.show()
```

# Findings
#### 1. Cancellations are low in Resort Hotels.
#### 2. Cancellations are high in City Hotels, along with high reservations.
#### 3. Ratio of not_cancelled to cancelled bookings is low for City hotels but it is high for Resort Hotels.
#### 4. Reason for this could be, the price of the Resort hotels is higher than the price of City Hotels. This is will be proved later in this project using ADR column.
#### 5. More focus should be laid on City hotels, along with more maintenance, in order to remove the reasons for cancellation of the city hotels booking.
#### 6. In resort Hotels, ADR might the major case of cancellations. But, in city hotels, maintenance and other facilities can be the reason of cancellation.

```python
resort_hotel = df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize= True)
```

```python
city_hotel = df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize= True)
```

# Findings

#### For City Hotels, 41% of the reservations are getting cancelled.

```python
# Checking effect of price on the cancellation of the City or Resort Hotels
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize= (20,8))
plt.title('Average Daily Rate (ADR) in City and Resort Hotel', fontsize = 30)
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel')
# x= index i.e., Reservation Status date and y = adr
plt.legend(fontsize = 20)
plt.show()
```

# Findings based the plot contructed above

#### 1. Spikes on the graph can be seen on the weekends, which represents that on weekends, the price of the City and Resorts Hotels is higher.
#### 2. Price of the City hotels is less in comparison to Resort hotels, for some of the days and for some of the days, the price of City Hotel is higher.
#### Note: Days can be weekends or holidays.

```python
# Different month has different number reservations.
# Hence checking, which month has the highest number of reservations and which one has highest number of cancellations.
```

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize = (16,8))
ax1 = sns.countplot(x = 'month', hue = 'is_canceled', data = df, palette = 'bright')
legend_labels,_ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor = (1,1))
plt.title('Reservation status per month', size = 20)
plt.xlabel('month')
plt.ylabel('number of reservations')
plt.legend(['not canceled', 'canceled'])
plt.show()
```

# Findings based on the plot:

#### 1. Highest number of cancellations occurred in January and lowest in August.
#### 2. Highest number of reservations occurred in August and lowest in December and January.

```python
plt.figure(figsize = (15,8))
plt.title('ADR per month', fontsize = 30)
sns.barplot(x= 'month', y='adr', data=df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())
plt.show()
```

# Findings

#### 1. ADR is lowest in August as per 'ADR per month' barplot. Also, as per 'Reservation status per month' graph, august has the least number of cancellations.

#### 2. ADR is highest in January as per 'ADR per month' barplot. Also, as per 'Reservation status per month' graph, January has the highest number of cancellations.

#### 2. This proves that when the price is higher, cancellation will also be higher. Reason being, as most of the people might not be able to afford the reservation rates

```python
# Finding, ADR is maximum for Cancelled or not cancelled reservation
```

```python
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace = True)
cancelled_df_adr.sort_values('reservation_status_date', inplace = True)

not_cancelled_data = df[df['is_canceled'] == 0]
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace = True)
not_cancelled_df_adr.sort_values('reservation_status_date', inplace = True)

plt.figure(figsize= (20,6))
plt.title('Average Daily Rate')
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label = 'not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label = 'cancelled')
plt.legend()
```

```python
#filtering data from 2016 to 2017 (Sep)
```

```python
cancelled_df_adr = cancelled_df_adr[(cancelled_df_adr['reservation_status_date'] > '2016')& (cancelled_df_adr['reservation_status_date']<'2017-09')]
not_cancelled_df_adr = not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date'] > '2016')& (not_cancelled_df_adr['reservation_status_date']<'2017-09')]

plt.figure(figsize= (20,6))
plt.title('Average Daily Rate', fontsize = 30)
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label = 'not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label = 'cancelled')
plt.legend(fontsize = 20)
plt.show()
```

### This clearly shows that price has clear affect on the cancellation rate of the reservations in the hotels. Thus, ADR is the factor, which is mostly influencing the cancellation rate.
